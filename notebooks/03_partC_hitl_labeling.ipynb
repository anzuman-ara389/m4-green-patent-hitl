In this part, I simulated Human-in-the-Loop learning. The uncertain samples selected in Part B were reviewed using LLM suggestions and human validation logic.

For each uncertain sample, I compared the LLM suggested label with the model prediction. If both agreed, I accepted the label automatically. If they disagreed, it simulated human correction by assigning the more reliable label.

These corrected labels are called HITL labels. They help improve label quality and reduce noise compared to silver labels.

This step demonstrates how combining human knowledge and machine learning improves data quality and model reliability.
